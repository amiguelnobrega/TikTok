```python
# what is the average view count of videos with 'claim' status?
claims = df[df['claim_status'] == 'claim']
print('Mean view count claims:', round(claims['video_view_count'].mean(), ndigits=2))
print('Median view count claims:', round(claims['video_view_count'].median(), ndigits=2))

#results show that there are no significant outlier values, mean and average are similar
```

```python
# what is the average view count of videos with 'opinion' status?
opinions = df[df['claim_status'] == 'opinion']
print('Mean view count opinions:', round(opinions['video_view_count'].mean(), ndigits=2))
print('Median view count opinions:', round(opinions['video_view_count'].median(), ndigits=2))

#results show that there are no significant outlier values, mean and average are similar
```

```python
#video breakdown by claim/opinion status

df.groupby(['claim_status', 'author_ban_status']).count()[['#']]

#claim videos appear to be more likely to be banned than opinion videos
```

```python
#breakdown of author_ban_status by video_view_count, video_like_count and video_share_count (mean and median)
df.groupby(['author_ban_status']).agg({
  'video_view_count': ['mean', 'median'],
  'video_like_count': ['mean', 'median'],
  'video_share_count': ['mean', 'median']})
```

```python
#video_share_count breakdown by author_ban_status
df.groupby(['author_ban_status']).median(numeric_only=True)[['video_share_count']]
```

```python
#breakdown of author_ban_status by video_view_count, video_like_count and video_share_count (count, mean and median)
df.groupby(['author_ban_status']).agg({
  'video_view_count': ['count', 'mean', 'median'],
  'video_like_count': ['count', 'mean', 'median'],
  'video_share_count': ['count', 'mean', 'median']})

#videos from banned and under review authors get far more engagement than active authors. the mean appears to be much higher in some groups confirming the presence of significant outliers
```

```python
#add a likes per view column to the dataset
df['likes_per_view'] = df['video_like_count'] / df['video_view_count']

#add a comments per view column to the dataset
df['comments_per_view'] = df['video_comment_count'] / df['video_view_count']

#add a shares per view column to the dataset
df['shares_per_view'] = df['video_share_count'] / df['video_view_count']
```

```python
#breakdown of author_ban_status by the above created columns
df.groupby(['author_ban_status', 'claim_status']).agg({
  'likes_per_view': ['count', 'mean', 'median'],
  'comments_per_view': ['count', 'mean', 'median'],
  'shares_per_view': ['count', 'mean', 'median']})

#
```

Of the 19,382 samples in this dataset, just under 50% are claims9,608 of them.
Engagement level is strongly correlated with claim status. This should be a focus of further inquiry.
Videos with banned authors have significantly higher engagement than videos with active authors. Videos with authors under review fall between these two categories in terms of engagement levels.

Videos by banned authors and those under review tend to get far more views, likes, and shares than videos by non-banned authors. However, when a video does get viewed, its engagement rate is less related to author ban status and more related to its claim status.

Also, we know that claim videos have a higher view rate than opinion videos, but this tells us that claim videos also have a higher rate of likes on average, so they are more favorably received as well. Furthermore, they receive more engagement via comments and shares than opinion videos.
